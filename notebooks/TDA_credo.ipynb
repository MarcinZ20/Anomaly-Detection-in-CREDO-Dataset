```python
import torch
from dataloader import TDA_Dataset
import matplotlib.pyplot as plt
```

```python
data = TDA_Dataset("images", True)
```

```python
x = 10

offset = 6809

plt.figure(figsize=(4, 4*x/2))

ind = 0
for index, (pers, img) in enumerate(data):

    if index >= x+offset:
        break
    if index < offset:
        continue

    ind = ind+1
    plt.subplot(x, 2, 2*ind + 1)
    plt.imshow(pers)
    plt.subplot(x, 2, 2*ind+2)
    plt.imshow(img)


```
